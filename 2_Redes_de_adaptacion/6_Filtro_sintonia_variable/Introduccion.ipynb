# Filtros sintonia variable    
 
Los filtros sintonízables pueden ser necesarios cuando los requisitos de selectividad requieren  de un número prohibitivamente grande de filtros fijos en un banco de filtros conmutados. 


Para ilustrar este punto, considere una situación en la que la frecuencia de la señal deseada se extiende de $f_{sigmin} = 100 MHz$ a $f_{sigmax} = 200 MHz$ y se desea atenuar el $2^{nd}$ armónico de la señal, de ese modo, la banda de detención abarca de $200 MHz$ a $400 MHz$. Para el ejemplo dado, suponga que el requisito mínimo de atenuación de banda de detención es de $40 dB$. 

Si el banco de filtros conmutados consiste  de filtros Butterworth pasa-bajo  
\footnote{Normalmente, los filtros elípticos o de Chebyshev se usarían para una mayor economía, pero los argumentos subyacentes son los mismos.}
, la atenuación de la banda de detención para un filtro específico viene dada por:

$$
\begin{equation}
A(f) = 10log_{10}[1+(\frac{f}{f_c})^{2N}]dB
\label{eq:1} \tag{1}
\end{equation}
$$

donde $f_c$ es la frecuencia de corte de $–3 dB$ del filtro y $N$ es el orden del filtro. 


La atenuación específica en la banda de frecuencia, $\eqref{eq:1}$ puede reescribirse como:

$$
\begin{equation}
\frac{f}{f_c} \le   [10^{0.1A(f)}-1]^{\frac{1}{2N}}
\label{eq:2}\tag{2}
\end{equation}
$$

donde el requisito de atenuación de frecuencia $f_{stop}$ viene dado por $A(f)$. 


Estas cantidades de filtro están relacionadas como se muestra en la Figura 1. 
Para suprimir adecuadamente el segundo armónico de la señal, debe ser cierto que 

$$
\begin{equation}
2f_{sig} \ge f_{stop}
\label{eq:3}\tag{3}
\end{equation}
$$

En consecuencia, el rango de frecuencia para un filtro pasa-bajos de grado $k^{th}$  viene dado por: 

$$
\begin{equation}
  [\frac{f_{stop}}{2} , f_{pass}]  
\label{eq:4}\tag{4}
\end{equation}
$$

## Bancos con filtros pasa-banda 
      
Alternativamente, se puede considerar un banco conmutado de filtros de paso de banda Butterworth. 
El número de filtros requeridos generalmente será menor si se usan filtros de pasa-banda, pero la complejidad del circuito sigue siendo bastante alta.  

En el caso del filtro pasa-banda Butterworth, \ref{eq:1} se puede aplicar siempre que la cantidad entre paréntesis se reemplace por la transformación de frecuencia de pasa-bajo a pasa-banda que está dada por:


$$
\begin{equation}
  u(f) = \frac{f^2 - f_{o}^2}{f B}   para k \ge 1 
\label{eq:6}
\end{equation}
$$

donde $f_o$ es la frecuencia central del filtro  y B es el ancho de banda del filtro ($–3 dB$ de ancho de banda en el caso de los filtros pasa-banda de Butterworth). 

Esta no es la única forma para tener un filtro pasa-banda, pero es muy conveniente para las discusiones en curso aquí. 
Los filtros de pasa-banda de esta forma exhiben simetría de frecuencia geométrica sobre la frecuencia central, pero se puede suponer que la simetría aritmética tiene poca pérdida de precisión siempre que el ancho de banda porcentual del filtro permanezca pequeño (típicamente $<20 \% $). 
Las frecuencias de corte correspondientes  del filtro:

$$
\begin{equation}
  f_{corte pasa-banda} = f_o \sqrt{1 + (\frac{B}{2 f_o})^2} \pm  \frac{B}{2}
\label{eq:7}
\end{equation}
$$

Para ilustrar la diferencia entre la simetría aritmética y la geométrica, las respuesta para un filtro pasa-banda Butterworth $5^\circ$-order que tiene un ancho de banda de $20\%$ se muestran en la Figura \ref{fig:3}. 
A medida que el ancho de banda de porcentaje se hace más pequeño, las dos respuestas en frecuencia se vuelve cada vez la misma como se muestra en la figura \ref{fig:4}.

El rango de frecuencia utilizable está dictado por el factor de forma del filtro en lo que respecta al requisito de atenuación de la banda de $A_{stop}$ y el límite de frecuencia inferior $fo B / 2$.  


Haciendo referencia a la figura 6, el ancho de banda de paso del filtro es B mientras que la anchura de banda de supresión en lo que respecta a la 2nd armónico de $f_o B / 2$ es $2 fo B$.

\end{equation} 
$$


Tenga en cuenta que el factor de forma de un filtro puede definirse de manera diferente en otras situaciones. 
El factor de forma dado por (8) se refiere solo a la segunda situación armónica de interés aquí. La cantidad $\frac{B}{f_o}$ también debe reconocerse como el ancho de banda fraccional del filtro.

La ecuación (6) debe considerarse como una transformación de frecuencia directa. 
Ya se ha dicho que el valor de la frecuencia transformada en los bordes de la banda de paso es (muy cercano) $\pm 1$. 
El valor de la frecuencia transformada en el $2^\circ$ valor armónico más bajo $(2 fo -B)$ viene dado por 


$$
\begin{equation}
  \gamma_{+} = u[2 (f_o - \frac{ B}{2})]
\label{eq:9}
\end{equation} 
$$

El valor de la frecuencia transformada en la región de la banda de parada de frecuencia más baja está dado de manera similar por 
$$
\begin{equation}
  \gamma_{-} = u[B]
\label{eq:10}
\end{equation} 
$$

Cuando el filtro pasa-banda se construye con base en \ref{eq:6}, resulta que la característica de atenuación de la banda de detención superior es un poco más floja que la característica de la banda de detención inferior. 
En consecuencia, siempre es cierto que $\gamma_{-} >\gamma_{+}$ . 
Por ejemplo, si $f_o = 100 MHz$ y $B = 10 MHz$, $\gamma_{+}= 13.737$ mientras que $\gamma_{-}= -99$. 
Esto es el resultado de la simetría geométrica subyacente de la característica de atenuación del filtro en lugar de ser simétrica aritmética como se discutió anteriormente con respecto a la Figura 3. 

Es conveniente definir un factor de forma del filtro basado en la relación de frecuencia transformada. 
Para este fin, el ancho de banda de frecuencia transformada es simplemente 2. 
Dado que el interés principal aquí se refiere a la supresión de armónicos de frecuencia, el ancho de banda de frecuencia transformada se considerará $2\gamma_{+}$. El factor de forma del filtro en términos de las cantidades de frecuencia transformadas es simplemente dado por $\frac{2\gamma_{+}}{2} = gamma_{+}$.

En base a este resultado junto con (1) y (6), la atenuación de la banda de detención asociada viene dada por 
$$
\begin{equation}
A(f) = 10log_{10}[1+(\gamma_{+})^{2N}]dB
\label{eq:11}
\end{equation}
$$
Continuando desde (9), no hay necesidad de llevar el subíndice "+" ya que solo la banda de detención superior será de interés a partir de este momento. La ecuación (9) se puede ampliar para dar 
$$
\begin{equation}
  (1+\gamma) B^2 - 2 f_o (2+\gamma) B - 3 f_o^2  = 0
\label{eq:12}
\end{equation} 
$$
Esta es una ecuación cuadrática simple que se puede resolver para el ancho de banda B dados los valores supuestos para la frecuencia central del filtro $f_o$ con el factor de forma nuevamente dictado por la Tabla 1 para una cantidad dada de atenuación de banda de parada requerida.

La solución se puede expresar en términos del ancho de banda fraccional n ($\chi$) dividiendo (12) por $f_o^2$ como 

$$
\begin{equation}
  (1+\gamma) \chi^2 - 2  (2+\gamma) \chi - 3   = 0
\label{eq:13}
\end{equation} 
$$
lo que lleva a 
$$
\begin{equation}
  \chi  = \frac{2+\gamma-\sqrt{\gamma^2+\gamma+1}}{1+\gamma}
\label{eq:14}
\end{equation} 
$$
Vale la pena comparar este resultado exacto con el resultado aproximado de (8). 
La ecuación (8) puede reformularse en términos de ancho de banda fraccional como 
$$
\begin{equation}
  \chi  = \frac{2}{2+factordeforma_{bpf}}
\label{eq:14}
\end{equation} 
$$
La relación aproximada entre el factor de forma del filtro y el ancho de banda fraccional viene dada por (8) (suponiendo la simetría del filtro aritmético) mientras que el resultado exacto está dado por (14). 
Estos resultados se muestran trazados juntos en la Figura 7. 
El resultado aproximado siempre es demasiado optimista en términos del ancho de banda permitido del filtro versus el resultado exacto. 
En el uso real, la respuesta de frecuencia real del filtro debe usarse para determinar cuántos filtros de paso de banda se requieren en el banco de filtros. 
Aun así, (14) es útil para fines de estimación.

Notas
 
Sin embargo, los filtros pasa-banda ajustables no siempre son la respuesta. Por ejemplo, 
 
    1.  Los bancos de filtros paso-bajo conmutados generalmente se requieren en aplicaciones de alta potencia porque la pérdida de inserción del filtro debe mantenerse lo más baja posible. 
	2.  Las corrientes de los inductores dentro de los filtros pasa-banda suelen escalar con el Q del valor filtro, lo que puede provocar el calentamiento del inductor. 
    3.  En aplicaciones de figura de ruido ultra bajo que requieren filtrado, los filtros LC pasa-bajo generalmente tendrán menos impacto en la figura de ruido que los filtros LC pasa-banda. 



# Sintonizable	de paso de banda	del filtro	Supuestos	y	Directrices

En los diseños de filtros LC tradicionales considerados en este documento, los únicos componentes físicos disponibles son resistencias, condensadores e inductores. En términos generales, es cierto que: 

    1. Los inductores son generalmente más grandes que los condensadores, especialmente a frecuencias más bajas. 
	2. Los inductores son casi siempre más caros que los capacitores. 
    3. Los inductores exhiben acoplamiento magnético no deseado si están situados demasiado cerca uno del otro. 
    4. El Q de los inductores son generalmente peores que el Q de los capacitores, lo que determina en gran medida la pérdida de inserción del filtro. 
    . Los bancos de inductores conmutados generalmente son de gran tamaño, los problemas de acoplamiento y el costo. 
    . Los inductores sintonizables generalmente no están disponibles, mientras que los capacitores sintonizables están disponibles en forma de diodos varactores. 
    7. Si se ajusta un capacitor frente a la frecuencia, generalmente  todos los capacitores en el filtro se deben ajustar. 
    8. Se asumen L y C ideales. 
    9. Las impedancias de los puertos de entrada y salida se supondrán iguales y constantes; típicamente $50 \Omega $. 

Dados estos puntos, es deseable (i) usar solo condensadores o varactores conmutados para los elementos sintonizables, y (ii) el número de inductores dentro de un filtro debe mantenerse al mínimo. 

 El nivel de impedancia dentro de un filtro cambia a medida que cambia la frecuencia central del filtro. Si este problema no se trata también, los problemas con la pérdida de inserción y la selectividad normalmente entrarán en juego. Considere el circuito RLC simple que se muestra en la Figura 11 que tiene un capacitor ajustable para cambiar la frecuencia de resonancia. La frecuencia de resonancia viene dada por   

$$
\begin{equation}
\omega_o = \frac{1}{\sqrt{CL}}
\label{eq:17}
\end{equation} 
$$

Como solo la capacitancia es ajustable, la frecuencia de resonancia es inversamente proporcional a $\sqrt{C}$, por lo que un cambio de frecuencia de $2:1$ requerirá un cambio de capacitancia de $4:1$.

El factor de calidad $Q$ del resonador viene dado por 

$$
\begin{equation}
Q = \frac{R}{\sqrt{2\pi f L}}
\label{eq:17}
\end{equation} 
$$

y es inversamente proporcional a la frecuencia central $f$, que generalmente no es deseable. 
Para tener un resonador $Q$ constante, la efectiva $R$ debe aumentar linealmente con la frecuencia. 
Si el ancho de banda del filtro debe permanecer constante con $f$ , el filtro $Q$ debe aumentar linealmente con $f$, lo que significa que la efectiva $R$ debe aumentar con $f^2$ . 
Estos y otros factores entran en juego al diseñar un filtro pasa-banda sintonizable.

Basado en (\ref{eq:25}), (\ref{eq:24}) y (\ref{eq:25}), la inductancia total de la serie viene dada por 

$$
\begin{equation}
L_s = L_1 + L_2 + L_3 = \frac{R_i}{\omega_c} \sqrt{\frac{R_s}{R_i}-1}+ \frac{Q_r R_i}{\omega_c} +\frac{R_i}{\omega_c} sqrt{\frac{R_l}{R_i}-1}
\label{eq:26}
\end{equation} 
$$

En el caso habitual donde $R_s = R_l$, (\ref{eq:26}) se puede simplificar a 

$$
\begin{equation}
L_s =  2 \frac{R_i}{\omega_c} \sqrt{\frac{R_s}{R_i}-1}+ \frac{Q_r R_i}{\omega_c}
\label{eq:27}
\end{equation} 
$$

Para poder implementar un filtro sintonizable, solo los capacitores de la Figura \ref{fig:14} deben ser ajustables y la inductancia serie L debe fijarse en el rango de interés de sintonía de frecuencia central. 
En consecuencia, (\ref{eq:27}) puede usarse para determinar $R_i$ con el fin de mantener a $L$ serie como una constante. Después de un procesamiento algebraico, (\ref{eq:27}) puede escribirse como  


$$
\begin{equation}
R_i^2 (\frac{4+Q_r^2}{\omega_c} )-R_i (\frac{2Q_r L_s}{\omega_c}+\frac{4R_s}{\omega_c^2} )+L_s^2 = 0 
\label{eq:28}
\end{equation} 
$$


Dado que el diseño requiere que $L_s$  sea constante en todo el rango de sintonización de frecuencia, idealmente la derivada de (\ref{eq:27}) igualada a cero debería ser en todo el rango de sintonía de interés. 
Sin embargo, no existe una solución, a menos que $R_i$ pueda cambiar con la frecuencia central .
