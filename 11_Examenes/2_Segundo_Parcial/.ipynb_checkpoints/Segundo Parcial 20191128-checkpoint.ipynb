```python
## Transformada discreta de Fourier

def myFourierPh(a, x, y, n):
    if a == 0 : 
        yp_o = 0.0
        for i,xi in enumerate(x):
            yp_o += y[i]*1/n
        y_p= yp_o
    else : 
        yp_ac = 0.0
        yp_as = 0.0
        for i,xi in enumerate(x):
            yp_ac += 2/n * y[i]*np.cos(a*xi)
            yp_as += 2/n * y[i]*np.sin(a*xi)
        y_p =  yp_ac + 1j*yp_as
    return y_p
```

```python
def vgsi(VGG,Vs,th):
    return VGG+Vs*cos(th)

def idi(v):
    i = 0.           if (v < 0.)  else 0
    i = (1.)*(v-0.) if (v >= 0   and v < 10) else i
    i = (10.)        if (v >= 10 ) else i
    return i
vgs = np.arange(-5.,20.,0.2)

ids = []
for v in vgs:
    ids.append(idi(v))
#%matplotlib inline
#plt.gca().cla()
plt.plot(vgs,ids,'ro')
#plt.axis([-10, 10, -1, 11])
plt.ylabel('ID')
plt.xlabel('VGS')
plt.rcParams['figure.figsize'] = (20.0, 4.0)
plt.grid()
plt.show()
```

### Potencia de DC

La componente de continua $i_{do}$ varia con el tiempo, por lo tanto tiene una componente de continua y una alterna de baja frecuencia. Para concer estas señales podemos emplear Fourier


$ I_{DCo} = \frac{1}{n} \sum_{i=1}^{n} i_{do}(\theta(i))  $

donde n es el número de muestras, en este caso 4.

$ I_{DCo} = \frac{1}{4} \cdot  ( 2.866 + 1.311 +0.001 + 1.311 ) =  1.372 A$

$ I_{DC1} = \frac{1}{n} \sum_{i=1}^{n} i_{do}(\theta(i)) \cdot \cos(\theta_m(i))  $

$ I_{DC1} = \frac{1}{n} \cdot  ( 2.866 \cdot \cos(0) + 1.311 \cdot \cos(90) +0.001 \cdot \cos(180) + 1.311 \cdot \cos(270) ) = 1.4327 A$


La potencia RMS de DC entonces :

$P_{DC} = VDD \cdot I_{DCo}  = 151.002W $

La potencia RMS de estas dos señales puede ser calculada como:

$P_{DC} = VDD \cdot I_{DCo} + VDD \cdot  \frac{I_{DC1}}{\sqrt{2}} = 151.002W + 111.438W = 262.44 W$


### Potencia de AM


La señal de AM esta compuesta por la componente de la poratadora, que este caso puede ser interpretada como la componente cero del espectro de Fourier de la corriente $i_{d1}$ y sus bandas laterales que corresponden a la fundamental (esta componente es la suma de ambas). 



$ I_{D1o} = \frac{1}{n} \sum_{i=1}^{n} i_{d1}(\theta_m(i))  $


donde n es el número de muestras, en este caso 4. Esta es la componete que corresponde a 

$ I_{D1o} = \frac{1}{4} (4.720 + 2.240 + 0.002 + 2.240) = 2.301 A   $


$ I_{D11} = \frac{1}{4} \sum_{i=1}^{4} i_{d1}(\theta_m(i)) \cdot cos(\theta(i)) $

$ I_{D11} = \frac{1}{4} (4.720 \cdot \cos(0)  + 2.240 \cdot \cos(90)  + 0.002 \cdot \cos(180)  + 2.240 \cdot \cos(270) ) = 1.179 A$



#### Potencia de salida 

El espectro de AM a la salida entonces esta compouesto por una portadora de valor $I_{ec} =  2.24A$, y sus bandas laterales estan compuesta por dos señales espaciadas por $1 KHz$ con valor de $I_{e1} =  1.17 A$.

Cuya potencia total en la carga es:

$P_o = \frac{i_d(10MHz)^2 \cdot R_c}{2} + \frac{i_d(10MHz-1KHz)^2 \cdot R_c}{2} + \frac{i_d(10MHz+1KHz)^2 \cdot R_c}{2} = 56.079W  + 14.733 W + 14.733 W= 85.545W $
