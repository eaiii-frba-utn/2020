###   Componente de continua $I_{dop}$ (condición de pico)

$I_{do} = \frac{1}{n} \sum_{i=1}^{n} idc_1(\theta_m(i))$

donde $n$ es el número de muestras. 

Se propone emplear $n = 24$, lo que corresponde a muestras cada $15$ grados.   

Por el metodo de los trapecios, primero obtenemos la tabla para 24 muestras.
La ecuación puede simplificarse a por su simetria par, teniendo en cuenta que en este caso, la primer muestra debe dividirse por 2.

$I_{do} = \frac{1}{12}  [ \frac{idc_1(\theta_m(1))}{2} + \sum_{i=2}^{\frac{n}{2}} idc_1(\theta_m(i)) ]$

Entonces, aplicando metodo de los trapecios:

$I_{do} = \frac{1}{12} [ \frac{i_d(0)}{2} +i_d(15) +i_d(30) +i_d(45)+ i_d(60) +i_d(75) + i_d(90)  ]$

Por el metodo de los trapecios, primero obtenemos la tabla para 24 muestras.

### Potencia de DC

La componente de continua $i_{do}$ varia con el tiempo, por lo tanto tiene una componente de continua y una alterna de baja frecuencia. Para concer estas señales podemos emplear Fourier


$I_{DCo} = \frac{1}{n} \sum_{i=1}^{n} i_{do}(\theta(i))$

donde n es el número de muestras, en este caso 4.

$I_{DCo} = \frac{1}{4} \cdot  ( 2.866 + 1.311 +0.001 + 1.311 ) =  1.372 A$

$I_{DC1} = \frac{1}{n} \sum_{i=1}^{n} i_{do}(\theta(i)) \cdot \cos(\theta_m(i))$

$I_{DC1} = \frac{1}{n} \cdot  ( 2.866 \cdot \cos(0) + 1.311 \cdot \cos(90) +0.001 \cdot \cos(180) + 1.311 \cdot \cos(270) ) = 1.4327 A$


La potencia RMS de DC entonces :

$P_{DC} = VDD \cdot I_{DCo}  = 151.002W$

La potencia RMS de estas dos señales puede ser calculada como:

$P_{DC} = VDD \cdot I_{DCo} + VDD \cdot  \frac{I_{DC1}}{\sqrt{2}} = 151.002W + 111.438W = 262.44 W$

$I_{D1o} = \frac{1}{n} \sum_{i=1}^{n} i_{d1}(\theta_m(i))$


donde n es el número de muestras, en este caso 4. Esta es la componete que corresponde a

# 8. En las condiciones del ítem 7, determinar $V_{GG}$  si se mantiene constante $V_s$.

La corriente de $id$ en la condición pico,

$i_{1p} = \frac{80V}{Rc} = \frac{80V}{21} = 3.77A$

En condición de pico $V_{sp} = V_{s} \cdot (1  +  m) = 12.087 V$


$$id(\theta) = \frac{1A}{1V}(V_{GG} + Vs \cos(\theta))$$
 
$$ i_{1p} = \frac{2}{\pi} \int_{0}^{\frac{\theta_c}{2}}   (V_{GG} + V_{sp} \cos(\theta)) \cdot \cos(\theta) d\theta$$

$$ i_{1p} = \frac{2}{\pi} \int_{0}^{\frac{\theta_c}{2}}   (V_{GG} \cdot \cos(\theta) + V_{sp} \cos(\theta) \cdot \cos(\theta))  d\theta$$


$$ i_{1p} = \frac{2}{\pi} \int_{0}^{\frac{\theta_c}{2}}   (V_{GG} \cdot \cos(\theta) + \frac{V_{sp}}{2} (1 + \cos(2\theta))  d\theta$$

$$ i_{1p} = \frac{2}{\pi} |  (V_{GG} \cdot \sin(\theta) + \frac{V_{sp}}{2} (\theta + \frac{\sin(2\theta)}{2})  |_{0}^{\frac{\theta_c}{2}}$$


$$ i_{1p} = \frac{2}{\pi} [  (V_{GG} \cdot \sin(\frac{\theta_c}{2}) + \frac{V_{sp}}{2} (\frac{\theta_c}{2} + \frac{\sin(2\frac{\theta_c}{2})}{2}) ] $$

$$ i_{1p} = \frac{2}{\pi} [  (V_{GG} \cdot \sin(\frac{\theta_c}{2}) + \frac{V_{sp}}{4} (\theta_c + \sin(\theta_c)) ]= 3.77A$$

$V_{GG} +  V_s \cdot cos(\frac{\theta_c}{2}) = 0V$

Entonces,

$$\frac{2}{\pi} [  -V_s \cdot cos(\frac{\theta_c}{2}) \cdot \sin(\frac{\theta_c}{2}) + \frac{V_{sp}}{4} (\theta_c + \sin(\theta_c)) ]= 3.77A$$

$$\cos(\frac{\theta_c}{2}) \cdot \sin(\frac{\theta_c}{2}) = \frac{\sin(\theta_c)}{2}$$



$$\frac{2}{\pi} [  (-V_s \cdot \frac{\sin(\theta_c)}{2} + \frac{V_{sp}}{4} (\theta_c + \sin(\theta_c) ]= 3.77A$$

$$\frac{V_{sp} \cdot 2}{\pi} [ \frac{\theta_c}{4}  + \frac{\sin(\theta_c)}{4} - \frac{\sin(\theta_c)}{2}   ]= 3.77A$$

$$\frac{V_{sp} \cdot 2}{\pi} [ \frac{\theta_c}{4}  - \frac{\sin(\theta_c)}{4}  ]= 3.77A$$

$$\frac{V_{sp}}{\pi} [ \theta_c  - \sin(\theta_c)   ]= 3.77A \cdot 2$$


$$\theta_c  - \sin(\theta_c)   =  \frac{3.77A \cdot 2 \cdot\pi}{V_{sp}}$$

$$\theta_c  - \sin(\theta_c)   =  1.9574$$

$$\theta_c = 2.53086$$

$$V_{GG} =  -3.6381 V$$

```python
## Transformada discreta de Fourier

def myFourierPh(a, x, y, n):
    if a == 0 : 
        yp_o = 0.0
        for i,xi in enumerate(x):
            yp_o += y[i]*1/n
        y_p= yp_o
    else : 
        yp_ac = 0.0
        yp_as = 0.0
        for i,xi in enumerate(x):
            yp_ac += 2/n * y[i]*np.cos(a*xi)
            yp_as += 2/n * y[i]*np.sin(a*xi)
        y_p =  yp_ac + 1j*yp_as
    return y_p
```

```python
def vgsi(VGG,Vs,th):
    return VGG+Vs*cos(th)

def vam(amp,m,vm,th):
    return amp*(1+m*vm)*np.cos(th)


def idi(v):
    i = 0.           if (v < 0.)  else 0
    i = (1.)*(v-0.) if (v >= 0   and v < 10) else i
    i = (10.)        if (v >= 10 ) else i
    return i
vgs = np.arange(-5.,20.,0.2)

ids = []
for v in vgs:
    ids.append(idi(v))
#%matplotlib inline
#plt.gca().cla()
plt.plot(vgs,ids,'ro')
#plt.axis([-10, 10, -1, 11])
plt.ylabel('ID')
plt.xlabel('VGS')
plt.rcParams['figure.figsize'] = (20.0, 4.0)
plt.grid()
plt.show()
```
