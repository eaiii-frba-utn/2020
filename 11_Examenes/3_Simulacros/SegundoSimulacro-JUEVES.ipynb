## 1. R1, R2, R3 y R5                                                                              1 punto


$$I_{cq}  = 5 mA$$

$$HFEq = 300$$

$$V_{BE}  = 0.6 V$$ 

$$V_{CE}  = 5 V$$

$$V_{CC}  = 12 V$$

### Calculos

$$I_b   = \frac{I_{cq}}{H_{FEq}}$$

$$I_{RB}  = I_b 10$$

$$V_{EE}  = V_{CC} - 2 V_{CE}$$

$$R_1   = \frac{(V_{EE}+V_{BE})}{I_{RB}} = 15600 \Omega$$

$$R_2   = \frac{V_{CE}}{I_{RB}} = 30000 \Omega$$

$$R_3   = \frac{(VCC- VCE)}{I_{RB}} = 42000 \Omega$$

$$R_5   = \frac{V_{EE}}{I_{cq}} = 400 \Omega$$

$R_1= 1.560e+04\Omega$

$R_2= 3.000e+04\Omega$

$R_3= 4.200e+04\Omega$

$R_5= 4.000e+02\Omega$


```python
import numpy as np

# Calculos de R1, R2, R3

Icq  = 5e-3

HFEq = 300

VBE  = 0.6 

VCE  = 5

VCC  = 12
# Calculos
Ib   = Icq/HFEq

IRB  = Ib *10

VEE  = VCC-2*VCE

R1   = (VEE+VBE)/IRB

R2   = VCE/IRB

R3   = (VCC- VCE)/IRB

R5   = VEE/Icq

print('$R_1= {:1.3e}\Omega$'.format(R1) )
print('$R_2= {:1.3e}\Omega$'.format(R2) )
print('$R_3= {:1.3e}\Omega$'.format(R3) )
print('$R_5= {:1.3e}\Omega$'.format(R5) )


```

## Q


$$|\bar{A_v}|^n =   \frac{1}{(1 + \chi_c^2)^\frac{n}{2}} = \frac{1}{\sqrt{2}}$$

$$(1 + \chi_c^2)^\frac{n}{2} = \sqrt{2}$$

$$(1 + \chi_c^2)^n = 2$$

$$\chi_c = \pm \sqrt{2^\frac{1}{n} - 1}$$

$$\chi= 0.64$$


El ancho de banda de n-etapas se calcula como

$$ Q = \frac{f_o}{BW_n} \sqrt{2^\frac{1}{n} - 1}$$

$$Q= 64.36$$

### Otra solucion

$$\chi = \frac{2 \cdot Q \cdot \Delta f}{f_o}$$

$$\frac{1}{1+ \chi^2} = \frac{1}{\sqrt{2}}$$

$$1+ \chi^2 = \sqrt{2}$$

$$(1+ \chi^2)^2 = 2$$

$$1 +2 \chi^2 + \chi^4 = 2$$

$$\chi^4 + 2 \chi^2 -1 = 0$$

$$\chi= 0.64$$

$$\chi = \frac{2 \cdot Q \cdot \Delta f}{f_o} $$

$$Q = \frac{1e6 171.57e-3}{2 10e3}$$

$$Q= 64.36$$


```python
# Calculo de X
fo = 1e6
BW = 10e3

# X^4+2X^2-1 = 0
c = -1
b =  2
a =  1

d = b**2-4*a*c
if d > 0: x = (-b+ (b**2-4*a*c)**0.5 )/(2*a)
else    : x = (-b+ (b**2+4*a*c)**0.5 )/(2*a)
print(x**0.5)

Qc = x**0.5 *fo/BW

print('$$\chi= {:1.2f}$$'.format(x**0.5))

print('$$Qc= {:1.2f}$$'.format(Qc))


```

## 3. L1, C1 y relación de espiras del trafo de entrada                   1 punto

$$C_1 = \frac{1}{L_1 \omega^2}$$

$$r_{ext} = 5000 \Omega$$

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_2}{r_{ext}}$$

$$L_2 = \frac{(\frac{1}{Q_c} - \frac{1}{Q_o}) \cdot r_{ext}}{\omega}$$

## 7. Por una falla en el circuito cambia el $Y_{12}$ del cascode pasando a valer $Y_{12}=-j1 \mu S$.   Calcular si el amplificador se vuelve inestable y justificar respuesta.  2 puntos


Linvil:

$$C = \frac{| y_{12} \cdot y_{21} | +\Re(y_{12} \cdot y_{21})}{2 \cdot \Re(y_{11}) \cdot \Re(y_{22})}$$

coeficiente de linvill $C= 0.25$, es incondicionalmente estable.

Stern para máxima transferencia de energía:  

$$K = \frac{2 \cdot \Re(y_{11} + y_{11}) \cdot \Re(y_{22}+ y_{22})}{| y_{12} \cdot y_{21} | +\Re(y_{12} \cdot y_{21})}$$

$$K= 16.0$$

El circuito es estable. 

 

```python
fo = 1e6
wo= 2*np.pi*fo
# Parametros del transistor 1 MHz
y11 =   1e-3 
y12 = - 1e-6 *1j
y21 =   50e-3  
y22 =   0.1e-3 


C = (abs(y12*y21)+(y12*y21).real)/(2*y11.real*y22.real)
print('###############  LINVILL   #####################')
print( 'coeficiente de linvill C= {:4.2F}' .format(C))
if C < 1:
    print( 'Es incondicionalmente estable')
else :
    print( 'Es potencialmente inestable')

K = (2*(y11.real*2)*(y22.real*2)) / (abs(y12*y21)+(y12*y21).real)
print('###############  STERN   #####################')
print( 'coeficiente de stern K= {:4.2F}' .format(K))
if K > 1:
    print( 'Estable')
else :
    print( 'Inestable')
```
