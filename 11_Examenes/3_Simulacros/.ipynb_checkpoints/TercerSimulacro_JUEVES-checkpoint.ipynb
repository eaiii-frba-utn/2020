## 1. Determinar el valor de Q1, Q2 y h 

### Estructura para máxima planicidad

Para lograr maxima planicidad en la banda de paso, se debe emplear una topologia Butterwoth. 

A continuación el diagrama de polos y ceros. 

![ex 2.13](fig/20200803_polos.png)

El simple sintonizado corresponde al polo que esta a frecuencia central, y los polos mas cercadon al eje $j\omega$ se logran con el doble sintonizado de la salida. 


### Simple sintonizado

$$Q_1 = \frac{f_o}{BW}$$

$$Q_1 = \frac{1MHz}{100kHz} = 10$$

$$Q_1= 10.00$$

### Doble sintonizado

$$Q_2 = \frac{fo}{BW \cdot \sin(30)} = \frac{1MHz}{100kHz \cdot \sin(30)}$$

$$Q_2= 20.00$$

$$k  = \frac{BW \cdot cos(30))}{fo} \frac{100kHz \cdot \cos(30)}{1MHz}$$

$$k = 8.66e-02$$

```python
import numpy as np

# Parametros del transistor 
r11   =   1/0.8e-3 
c11   =   10e-12
r22   =   1/0.1e-3 
c22   =   10e-12

rg    = 1250
rl    =   2.5e3
N1    =   100
# Calculos auxiliares
####### 1
BW = 100e3
fo = 1e6
wo = (2*np.pi*fo)
# Simple sintofizado
Q1 = fo/BW

# Doble 

Q2 = fo/(BW*np.sin(30*np.pi/180))

k  = (BW*np.cos(30*np.pi/180))/fo

print('$$Q_1= {:1.2f}$$'.format(Q1))

print('$$Q_2= {:1.2f}$$'.format(Q2))

print('$$k = {:1.2e}$$'.format(k))
```

## 2. Determinar $C_1$, $C_2$, $L_1$, $N_2$



### $L_1$

Para el calculo del inductor $L_1$, es importante notar la condicion $L_1 = L_2 = L_3$. 

El inductor $L_2$ es el unico que tiene impuesta la carga, por lo tanto, es necesario realizar el calculo de este componente primero. 

### Calculo del inductor $L_2$


![ex 2.13](fig/20200803_circuito_out.png)

Ese inductor corresponde a uno de los sintonizado del doble sintonizado y tiene como carga la impedancia de salida del transistor.

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_2}{r_{ext}}$$

Al ser un doble sintonizado, la resistencia de carga del sintonizado corresponde al parte resistiva de la admtancia de salida del transistor.

$$r_{ext} = r_{22}$$

El calculo de $Q_c$ para el sintonizado conformado por $L_2$

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_2}{r_{22}}$$

Despejando $L_2$

$$L_2 = (\frac{1}{Q_c} - \frac{1}{Q_o}) \cdot \frac{r_{22}}{\omega}$$

$$L_2 = (\frac{1}{20} - \frac{1}{\inf}) \cdot \frac{10k\Omega}{2 \pi 1e6MHz}$$

$$L_2 = 79.577e-06 Hy$$

$$L_1 = L_2 = 7.9577e-05 Hy$$

```python
# DOBLE SINTONIZADO
#  calculo a la salida sobre COLECTOR

Qo = np.inf

L2 = (1/Q2-1/Qo)*r22/wo

L1 = L2

print('$$L2= {:1.2e} Hy$$'.format(L2))

print('$$L1= {:1.2e} F$$'.format(L1))
```

### Calculo del transformador de entrada

![ex 2.13](fig/20200803_circuito_in.png)

Definido el inductor, es momento de calcular la relación de transformación. 

Teniendo valor del inductor, debemos conocer la resistencia de tope de bobina para lograr el $Q_c$ necesario para el simple sintonizado. 
La resistencia externa total depende del $Q_c$ del simple sintonizado. 

$$\frac{1}{Q_c} = \frac{1}{Q_o} + \frac{\omega L_1}{r_{ext}}$$

$$r_{ext} = \frac{\omega L_2}{(\frac{1}{Q_c} - \frac{1}{Q_o})}$$


$$rext = 5000.000 \Omega$$

Para MTEQcte $r'_g=r'_{11}$

$$r_{ext} = \frac{r'_g \cdot r'_{11}}{r'_g + r'_{11}}$$


El calculo de la relación del transformador

$$\frac{r_{11}}{r'_{11}} = (\frac{N_2}{N_1})^2$$ 

$$N_2 = \sqrt{\frac{r_{11}}{r'_{11}}} \cdot N_1$$ 

$$N_2= 35.36$$

```python
# ENTRADA!
L1 = L2
rext = wo*L1/(1/Q1-1/Qo)
print('$$rext = {:1.3f} \Omega$$'.format(rext))
rt   = rext*2

print('$$rt = {:1.3f} \Omega$$'.format(rt))

# (N1/N2)**2 = rt/r11
N2 = (r11/rt)**0.5*N1

print('$$N2= {:1.2f}$$'.format(N2))

L12 = L1*(N2/N1)**2
```

### Calculo del divisor capacitivo

El capacitor de sintonia para el inductor

$$C_s = \frac{1}{L_1 \cdot \omega^2}$$

$$C_s= 3.18e-10 F$$

Esta conformado por la capacidad de entrada  del transistor $c_{11}$ y la capacidad del divisor capacitivo, ambos visto a tope de bobina $L_1$.

$$c_s = c'_{11} + c_s'$$

$$c'_{11} = (\frac{N_2}{N_1})^2 \cdot c_{11}$$

$$c'_{11}= 1.25e-12 F$$


![ex 2.13](fig/20200803_circuito_divC_in.png)

#### Divisor capacitivo

Capacitor a tope de bobina

$$c'_s = c_s - c'_{11}$$

$$c'_s = 315e-12 F$$


#### Relación del divisor

$$Nc1 =\sqrt{\frac{r'_g}{r_g}}$$


$$Nc1 = 2.83$$


#### Calculo del divisor

$$Q_{m2} = r'_g \omega C'_s$$


$$Qm2 = 19.69$$

Como $Q_{m2} > 10$ 

$$Q_{m1} \sim \frac{Q_{m2}}{Nc1}$$

$$Q_{m1} \sim 7.11$$


Como $Q_{m1} < 10$ entonces: 

Volvemos a calcular $Q_{m1}$: 

$$Q_{m1} = \sqrt[]{\frac{(1+Q_{m2}^2)}{N^2}-1}$$

$$Qm1 = 6.90$$

$$C_1 = \frac{Q_{m1}}{r_g \omega}$$ 

$$C_1= 878e-12 F$$


$$C'_1 = C_1(1+\frac{1}{Q_{m1}^2})$$

$$C     = \frac{C'_s}{ (1+\frac{1}{Q_{m2}^2})}$$

$$C_2 = \frac{C'_1 C}{C'_1 - C}$$

$$C_2= 480e-12 F$$


### Si ambos son mayores a 10

$$C_{2a} = c'_s \frac{Nc1}{Nc1-1}$$

$$C_1a= 896.8e-12 F$$


```python
CS = 1/(L1*wo**2) 
print('$$C_S= {:1.2e} F$$'.format(CS))

c11t = c11/((N1/N2)**2)
print('$$C_11t= {:1.2e} F$$'.format(c11t))

CSi = CS - c11t
print('$$C_Si= {:1.2e} F$$'.format(CSi))


# Divisor Capacitivo transistor
# R//C == --C1-C2//Ro
R = rt
C = CSi 
print(C)
Ro= rg

# (ro/ri) = (nc)^2
nc1 =  (R/Ro)**(1/2) 
print('$$Nc1 = {:1.2f} $$'.format(nc1))

Qm2 = wo*C*R 
print('$$Qm2 = {:1.2f} $$'.format(Qm2))

Qm1 = ( ((1+Qm2**2)/nc1**2) - 1)**0.5
print('$$Qm1 = {:1.2f} $$'.format(Qm1))

# para el calculo de los capacitorees
C1  = Qm1/(Ro*wo)

C1p = C1*(1+1/Qm1**2)

Cp  = C /(1+1/Qm2**2)

C2  = C1p*Cp/(C1p-Cp)

print('$$C_1= {:1.3e} F$$'.format(C1))
print('$$C_2= {:1.3e} F$$'.format(C2))

C2a = nc1/(nc1-1)*C
C1a = nc1*C
print('$$C_1a= {:1.3e} F$$'.format(C1a))
print('$$C_2a= {:1.3e} F$$'.format(C2a))
```


## 4. Determinar el rechazo de Frecuencia Imagen del Rx

## Butterwoth de tercer orden

La transferencia normalizada de un Butterwoth de orden $n$ puede expresarse como: 

$$\bar{|H(f_{c})|}=\frac{1}{\sqrt{1+(\frac{\chi(f)}{\chi(f_c)})^{2\cdot n}}}$$ 

donde:

$$\chi(f)   = 2 \cdot Q \cdot \frac{(f-f_o)}{f_o}$$

$$\chi(f_c) = 2 \cdot Q \cdot \frac{(f_c-f_o)}{f_o}$$

$$n = 3$$

$$\bar{|H(f_{c})|}=\frac{1}{\sqrt{1+(\frac{(f-f_o)}{(f_c-f_o)})^{6}}}$$ 

$$|\bar{A_v(img)} | = 0.00016$$

$$|\bar{A_v(img)} | = -75.60dB$$



```python
fif = 455e3
fosc = fo + fif

fimg = fo + 2 * fif
print(fimg)
Avimg = 1/(1+( (fimg-fo)/(BW/2)) **6)**0.5

print('$$|  A_v(img) | = {:1.4f}$$'.format(Avimg) )

print('$$|  A_v(img) | = {:1.2f}dB$$'.format(20*np.log10(Avimg)) )
```

## 5. Determinar la ganancia de potencia (G=Pout/Pin) del amplificador de RF en frecuencia central


### Calculo de $g_m$

Antes de calcular la ganancia de potencia, debemos conocer $g_m$. 

$$g_m = 40 \cdot I_{cq}$$

Para calcular $I_{cq}$, debemos calcular la tensión $V_{BB}$. 

Suponiendo que la corriente $I_B$ es despresiable a la que circula por el divisor resistivos (luego debemos demostrarlo).

$$V_{BB} = V_{CC} \frac{R_2}{R_1+R_2}$$



$$V_BB=4.00e+00 V$$

$$V_E  = V_{BB}-V_{BE}$$

$$V_E=3.40e+00 V$$

$$I_E  = \frac{V_E}{R_E}$$

$$I_E=2.83e-03A$$


$$I_C  = \frac{I_E}{(1+\frac{1}{H_{FE}})}$$

$$I_C=2.82e-03A$$

$$I_B  = \frac{I_C}{HFE}$$

$$I_B=9.41e-06A$$

Verificando entonces.

$$I_{RB} = \frac{V_{BB}}{R_2}$$

$$I_{RB} = 222e-06A$$

Que es mucho mayor a la corriente de la base. 

$$g_m = 40 \cdot I_C$$

$$gm = 113e-03S$$

### Ganancia de potencia
La ganacia de potencia se calcula como:
$$G_{1} = (\frac{v_o}{v_i})^2 \frac{r_i}{r_o}$$

Por lo que es necesario calcular la ganacia de tensión:

$$| \bar{Av} | = \frac{v_o}{v_i}$$

La ganancia de tensión se se puede calcular mediante las ganancias parciales. 

$$| \bar{Av} | = \frac{v_o}{v_c} \cdot \frac{v_c}{v_b} \cdot \frac{v_b}{v_a}  \cdot \frac{v_a}{v_i}$$



$$\frac{v_o}{v_c} = \frac{1}{N_{c2}}$$

$$|\frac{v_o}{v_c}| = 0.50$$



$$|\frac{v_c}{v_b}| =  gm L_2 w_o Q_2\frac{h}{(1+h^2) }$$ 


$$|\frac{v_c}{v_b}| = 489.12$$




$$\frac{v_b}{v_a} =\frac{N_2}{N_1}$$

$$|\frac{v_b}{v_a}| = 0.35$$

$$\frac{v_a}{v_i} = N_{c1}$$

$$|\frac{v_a}{v_i}| = 2.83$$

$$| \bar{Av} | = \frac{v_o}{v_c} \cdot \frac{v_c}{v_b} \cdot \frac{v_b}{v_a}  \cdot \frac{v_a}{v_i}$$

$$|Avo| = 244.56$$
$$|Avo| = 47.77dB$$


$$G_{1} = (\frac{v_o}{v_i})^2 \frac{r_i}{r_o}$$

$$|G_1| = 29904.47$$
$$|G_1| =44.76dB$$


```python
R1  = 36e3
R2  = 18e3
RE  = 1.2e3
VCC = 12
HFE = 300 
VBE = 0.6

VBB = VCC/(R1+R2)*R2
print('$$V_BB={:1.2e} V$$'.format(VBB) )

VE  = VBB-VBE
print('$$V_E={:1.2e} V$$'.format(VE) )

IE  = VE/RE
print('$$I_E={:1.2e}A$$'.format(IE) )


IC  = IE/(1+1/HFE)
IB  = IC/HFE
print('$$I_C={:1.2e}A$$'.format(IC) )
print('$$I_B={:1.2e}A$$'.format(IB) )


print('$$IRB = {:1.2e}A$$'.format(VBB/R2) )

gm = 40*IC
print('gm = {:1.2e}S'.format(gm) )

h = k * Q2
M = k*L2


Avdso = gm*wo*L2*Q2*h/(1+h**2)
print(Avdso)
print('L2 = {:1.2e}Hy'.format(L2) )

print('gm = {:1.2e}S'.format(gm) )
print('fo = {:1.2e}Hz'.format(wo/(2*np.pi)) )
print('Q2 = {:1.2e}'.format(Q2) )
print('h = {:1.2e}'.format(h) )



print('$$|\frac v_o v_c| = {:1.2f}'.format(1/nc2))
    
print('$$|\frac v_c v_b| = {:1.2f}'.format( Avdso))

print('$$|\frac v_b v_a| = {:1.2f}'.format(N2/N1))

print('$$|\frac v_a v_i| = {:1.2f}'.format(nc1))

   


Avo = nc1*N2/N1 * Avdso * 1/nc2


print('$$|Avo| = {:1.2f}'.format(Avo))
print('$$|Avo| = {:1.2f}dB'.format(20*np.log10(Avo)))


G1 = Avo**2 *rg *1/rl

print('$$|G_1| = {:1.2f}'.format(G1))
print('$$|G_1| = {:1.2f}dB'.format(10*np.log10(G1)))
```

## 6. Si se cambian los sintonizados de salida por un único sintonizado, manteniendo las condiciones de adaptación, y el elemento activo se reemplaza por otro con los mismos parámetros admitancia excepto en el parámetro y12 que pasa a valer $y_{12}=-j w_o 10^{-12} S$, determinar, justificando la respuesta, si el circuito es estable


Linvil:

$$C = \frac{| y_{12} \cdot y_{21} | +\Re(y_{12} \cdot y_{21})}{2 \cdot \Re(y_{11}) \cdot \Re(y_{22})}$$

coeficiente de linvill $C= 4.45$, es potencialmente inestable.

Stern para máxima transferencia de energía:  

$$K = \frac{2 \cdot \Re(y_{11} + y_{11}) \cdot \Re(y_{22}+ y_{22})}{| y_{12} \cdot y_{21} | +\Re(y_{12} \cdot y_{21})}$$

$$K= 0.90$$

El circuito es inestable. 

```python
fo = 1e6
wo= 2*np.pi*fo
# Parametros del transistor 120 MHz
y11 =   .8e-3 
y12 = - 1e-12*wo *1j
y21 =   gm  
y22 =   0.1e-3 


C = (abs(y12*y21)+(y12*y21).real)/(2*y11.real*y22.real)
print('###############  LINVILL   #####################')
print( 'coeficiente de linvill C= {:4.2F}' .format(C))
if C < 1:
    print( 'Es incondicionalmente estable')
else :
    print( 'Es potencialmente inestable')

K = (2*(y11.real*2)*(y22.real*2)) / (abs(y12*y21)+(y12*y21).real)
print('###############  STERN   #####################')
print( 'coeficiente de stern K= {:4.2F}' .format(K))
if K > 1:
    print( 'Estable')
else :
    print( 'Inestable')

```

## 7. Sabiendo que el resto del Rx (Mezclador, FI, Detect, Audio) tiene una NF=15 dB y el ancho de banda de ruido Ben=1 kHz está dado por un filtro a la salida de la FI determinar el nivel de señal en bornes de antena (V1). 


La cifra de ruido total del receptor se calcula como:
$$F_{tv}=F_{1v}+\frac{F_{2v}- 1}{G_{1v}}$$

Los valores deben estar en veces:

$$F_{1} = 3dB$$
$$F_{1v} = 1.995$$
$$G_1  = 30104.167$$ 

$$F_{2} = 15dB$$
$$F_{2v} = 31.623$$

$$F_{tv}=F_{1v}+\frac{F_{2v}- 1}{G_{1v}}$$

$$F_{tv} = 1.996$$

La cifra de ruido es la relación de señal a ruido de la entrada respecto a la salida:
$$F_{tv}= \frac{SNR_i}{SNR_o}$$

$$SNR_i = SNR_o \cdot F_{tv}$$

$$SNR_i = 19.96e+00$$

La potencia de ruido a entrada, dada la antena a $T_e = 300 K$, para un ancho de banda equivalente de ruido de $B_{qe} = 1KHz$:

$$N_i = KTB_{eq}$$


$$N_i = 1.38e−23 300 1e3$$

$$N_i = 4.140e-18 W$$

$$SNR_i   = \frac{P_i}{N_i}$$

$$P_i = SNR_i \cdot N_i$$

$$P_i = 8.262e-17 W$$

$$V_i = \sqrt{P_i \cdot r_g}$$

$$V_i = 3.214e-07 V$$


```python
# A la salida del Rx (sobre el parlante) hay una señal alterna de 1V y SNR = 10 dB.
F1=3
F2=15
SNR0d=10
Tant = 300
BEN  = 1e3
Vo   = 1

F1v = 10**(F1/10)
F2v = 10**(F2/10)

Ftv  = F1v + (F2-1)/G1
print('$$F1v = {:1.3f}$$'.format(F1v))
print('$$G1  = {:1.3f}$$'.format(G1))
print('$$F2v = {:1.3f}$$'.format(F2v))
print('$$Ftv = {:1.3f}$$'.format(Ftv))


SNRo = 10**(SNR0d/10)
print('$$SNRo = {:1.3e} V$$'.format(SNRo))


SNRi = SNRo*Ftv
print('$$SNR_i = {:1.3e}$$'.format(SNRi))

Ni   = 1.38e-23 * Tant * BEN
print('$$N_i = {:1.3e} W$$'.format(Ni))

Pi   = Ni * SNRi
print('$$P_i = {:1.3e} W$$'.format(Pi))

Vi   = (Pi*rg)**0.5
print('$$V_i = {:1.3e} V$$'.format(Vi))
```

## 8. Determinar el ancho de banda del Rx sabiendo que la etapa de FI está conformada por 3 Simple Sintonizados sincrónicos

Los canales adyacentes se encuentran a $+/- 50 kHz$. 
Se especifica un rechazo mínimo a los mismos de $44 dB$

$$|\bar{A_{3ss}}| = \frac{1}{(1+Q_{if}^2 (\frac{f_{ca}}{f_{if}} - \frac{f_{if}}{f_{ca}})^2 )^{\frac{3}{2}}} = \frac{1}{10^{\frac{Attca}{10}}}$$

$$Q_{if} = \frac{\sqrt{(10^{\frac{Attca}{10}})^{\frac{2}{3}} - 1}}{(\frac{f_{ca}}{f_{if}} - \frac{f_{if}}{f_{ca}})}$$

$$Q_if = 25.460$$

$$BW_{1ss} = \frac{f_o}{Q_if}$$

$$BW_{3ss} = BW_{1ss} \sqrt{2^{\frac{1}{3}}-1}$$

$$BW_{3ss}  = 9111.322 Hz$$

```python
#  Los canales adyacentes se encuentran a +/- 50 kHz. Se especifica un rechazo mínimo a los mismos de 44 dB

ATTca = 44 #dB
fif   = 455e3
fca   = 50e3
fci   = fif + fca
n     = 3

Qif = (((10**(ATTca/20))**(2/n))-1)**(0.5) /(fci/fif-fif/fci)

BWift = (fif/Qif) * (2**(1/n)-1)**0.5

print('$$Q_if = {:2.3f}$$'.format(Qif))
print('$$BWif_t = {:2.3f} Hz$$'.format(BWift))
```

## 9. Al Rx ingresa una señal de AM modulada por un tono de audio de 5 kHz con m=80%. Determinar el índice de modulación a la entrada del detector


En una señal de AM con un indice de modulación de  $m= 80\%$, corresponde a bandas laterales ($v_{bli}$ y $v_{bls}$) con una amplitud del $40\%$ de la portadora ($v_{c}$). 


$$m_i = 0.80$$


$$|\frac{v_{bls}}{v_{c}}| = \bar{v_{bls}} = \frac{m_i}{2}$$

$$\bar{v_{bli}} = 0.4$$

$$|\bar{v_{c}}| = 1.0$$

$$\bar{v_{bls}} = \frac{m_i}{2}$$

$$\bar{v_{bls}} = 0.4$$

Las frecuencias de las bandas laterales estan a $5 KHz$ de la frecuencia de la portadora.

$$f_{bli} = f_{if} - 5KHz$$

$$f_{bli} = 440 KHz$$

$$f_{c} = f_{if}$$

$$f_{c} = 455 KHz$$

$$f_{bls} = f_{if}+ 5KHz$$

$$f_{bls} = 450 KHz$$ 

Esta señal es filtrada por la etapa de IF, cualculada en el punto anterior. 

$$|\bar{A_{3ss}}(f)| = \frac{1}{(1+Q_{if}^2 (\frac{f}{f_{if}} - \frac{f_{if}}{f})^2 )^{\frac{3}{2}}}$$

$$|\bar{A_{3ss}}(f_{bli})| = 0.662$$

$$|\bar{A_{3ss}}(f_{c})|   = 1.0$$

$$|\bar{A_{3ss}}(f_{bls})| = 0.667$$

$$\bar{v_{blso}} = |\bar{A_{3ss}}(f_{bls})| \cdot \bar{v_{bls}}$$

$$\bar{v_{blio}} = 0.27$$

$$\bar{v_{co}}   = 1.00$$

$$\bar{v_{blso}} = 0.27$$

$$v_{max} =(\bar{v_{co}}+ \bar{v_{blso}} + \bar{v_{blio}})$$

$$v_{min} =(\bar{v_{co}}- \bar{v_{blso}} - \bar{v_{blio}})$$

$$m_o = \frac{v_{max}- v_{min}}{v_{max}+ v_{min}}$$ 

$$m_o = 0.53$$


```python
mi   = 0.8
fif  = 455e3 
df   = 5e3



vbli  = mi/2
fbli  = fif - df
Abli  = 1/(1+Qif**2*(fbli/fif-fif/fbli)**2)**(3/2)
print('$$Abli  = {:1.3f}$$'.format(Abli))
vblio = vbli*Abli
print('$$vblso = {:1.2f}$$'.format(vblio))

vbls  = mi/2
fbls  = fif + df
Abls  = 1/(1+Qif**2*(fbls/fif-fif/fbls)**2)**(3/2)
print('$$Abls  = {:1.3f}$$'.format(Abls))
vblso = vbls*Abls
print('$$vblso = {:1.2f}$$'.format(vblso))




mo = vblso*2
print('$$mo = {:1.2f}$$'.format(mo))
```

## 10. Por una falla en el OL el mismo corre su frecuencia a 1460 kHz. Obtener el índice de modulación a la entrada del detector


$$|\frac{v_{bls}}{v_{c}}| = \bar{v_{bls}} = \frac{m_i}{2}$$

$$\bar{v_{bli}} = 0.4$$

$$|\bar{v_{c}}| = 1.0$$

$$\bar{v_{bls}} = \frac{m_i}{2}$$

$$\bar{v_{bls}} = 0.4$$

Las frecuencias de las bandas laterales estan a $5 KHz$ de la frecuencia de la portadora.

$$f_{bli} = f_{if} - 5KHz + 5KHz$$

$$f_{bli} = 445 KHz$$

$$f_{c} = f_{if} + 5KHz$$

$$f_{c} = 450 KHz$$

$$f_{bls} = f_{if}+ 5KHz + 5KHz$$

$$f_{bls} = 455 KHz$$ 


Esta señal es filtrada por la etapa de IF, cualculada en el punto anterior. 

$$|\bar{A_{3ss}}(f)| = \frac{1}{(1+Q_{if}^2 (\frac{f}{f_{if}} - \frac{f_{if}}{f})^2 )^{\frac{3}{2}}}$$

$$|\bar{A_{3ss}}(f_{bli})| = 1.0$$

$$|\bar{A_{3ss}}(f_{c})|   = 0.667$$

$$|\bar{A_{3ss}}(f_{bls})| = 0.301$$


$$\bar{v_{blso}} = |\bar{A_{3ss}}(f_{bls})| \cdot \bar{v_{bls}}$$

$$\bar{v_{blio}} = 0.40$$

$$\bar{v_{co}}   = 0.67$$

$$\bar{v_{blso}} = 0.12$$

$$v_{max} =(\bar{v_{co}}+ \bar{v_{blso}} + \bar{v_{blio}})$$

$$v_{min} =(\bar{v_{co}}- \bar{v_{blso}} - \bar{v_{blio}})$$

$$m_o = \frac{v_{max}- v_{min}}{v_{max}+ v_{min}}$$ 

$$mo = 0.78$$


```python
mi   = 0.8
fif  = 455e3 
df   = 5e3
dfo  = 5e3


vbli  = mi/2
fbli  = fif - df + dfo
Abli  = 1/(1+Qif**2*(fbli/fif-fif/fbli)**2)**(3/2)
print('$$Abli  = {:1.3f}$$'.format(Abli))
vblio = vbli*Abli
print('$$vblso = {:1.2f}$$'.format(vblso))


vc  = 1
fc  = fif  + dfo
Ac  = 1/(1+Qif**2*(fc/fif-fif/fc)**2)**(3/2)
print('$$Ac  = {:1.3f}$$'.format(Ac))
vco = vc*Ac
print('$$vco = {:1.2f}$$'.format(vco))


vbls  = mi/2
fbls  = fif + df + dfo
Abls  = 1/(1+Qif**2*(fbls/fif-fif/fbls)**2)**(3/2)
print('$$Abls  = {:1.3f}$$'.format(Abls))
vblso = vbls*Abls
print('$$vblso = {:1.2f}$$'.format(vblso))


vmax = vco + vblio + vblso
vmin = vco - vblio - vblso

mo = (vmax-vmin)/(vmax+vmin)
print('$$mo = {:1.2f}$$'.format(mo))


print(mo)
```
