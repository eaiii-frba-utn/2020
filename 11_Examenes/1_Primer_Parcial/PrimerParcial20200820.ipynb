## Punto 1 

Para la etapa de RF, el valor de los componentes de las redes de adaptación, siendo los datos del transistor:  $y_{11}= (1+j0)\,mS$; $y_{22}= (0.1+j0)\,mS$; $y_{21}= (10+j0)\,mS$; $y_{12}=0$ 
 

Del enunciado, la frecuencia imagen se atenua $40 dB$. 


Para la etapa de RF, $Qc$ de los sintonizados para cumplir con la condición de filtros sincrónicos y $Q_{c2} = 2 \cdot Q_{c1}$  .


$\frac{1}{\sqrt{1+Q_1^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2}} \cdot \frac{1}{\sqrt{1+Q_2^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2}} = \frac{1}{10^{\frac{40}{20}}}$


$\frac{1}{\sqrt{1+Q_1^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2}} \cdot \frac{1}{\sqrt{1+(2\cdot Q_1)^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2}} = \frac{1}{10^{\frac{40}{20}}}$

$(1+Q_1^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2) \cdot (1+ 4\cdot Q_1^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2) = 100$

$1 + 5\cdot Q_1^2\cdot(\frac{fi}{fo}-\frac{fo}{fi})^2 + 4\cdot Q_1^4\cdot(\frac{fi}{fo}-\frac{fo}{fi})^4 = 100$

$f_i = f_{osc} - f_{1FI} = 109.7 MHz - 10.7 MHz = 98.6 MHz$

$(\frac{fi}{fo}-\frac{fo}{fi})^2 = (\frac{98.6}{120}-\frac{120}{98.6})^2 = 0.1563$

$-99 + 5\cdot Q_1^2\cdot(0.1563) + 4\cdot Q_1^4\cdot(0.1563)^2 = 0$

$$Q_1 =  5.30$$
$$Q_2 =  10.60$$



### Sintonizado salida

$\frac{1}{Q2} = \frac{1}{Qo} + \frac{\omega L}{r_{ext}}$
$$L= 5.93e-07 Hy$$

$C_s = \frac{1}{L \cdot \omega^2}$

$$C_s= 2.97e-12 F$$

$Q_m  = \sqrt{(\frac{r22}{rl}) - 1}$

$C_4  = \frac{1}{rl \omega_o Q_m}$

$$C_4= 4.42e-13 F$$

$C_{4p} = \frac{C_4}{(1+\frac{1}{Q_m^2})}$

$C_3  = C_s - C_{4p}$

$$C_3= 2.57e-12 F$$

### Sintonizado entrada

$r_{ext1} = \frac{L \omega_o}{(\frac{1}{Q_1}-\frac{1}{Q_o})}$

$r_{11t}  = rext1 \cdot 2$

$r_{gt}   = rext1 \cdot 2$

$Q_{m2}   = r_{11t} \omega_o C_s$

$$Q_{m2} = 10.89$$

$N_c      = \sqrt{\frac{r_{11t}}{r_{11}}}$

$Q_{m1}   = \frac{Q_{m2}}{N_c}$

$$Q_{m1} = 4.94$$


$C_2  = C_s \cdot  N_c$
$$C_2= 6.55e-12 F$$

$C_1  = \frac{C_s N_c}{(N_c-1)}$

$$C_1= 5.43e-12 F$$

$N_1 = N \cdot \sqrt{\frac{r_g}{r_{gt}}}$


$$N_1 = 2.03$$



```python
import numpy as np

# Parametros del transistor 
r11   =   1/1e-3 
r22   =   1/0.1e-3 
gm    =   10e-3

rg    = 50
rl    = 1e3
N     = 20
n     = 2 
# Calculos auxiliares
####### 1
att  = 40
fo   = 120e6
wo   = (2*np.pi*fo)
fif  = 10.7e6
fimg = fo-2*fif
print(fimg)

####
df = ((fimg/fo)-(fo/fimg))**2
print(df)
c = -((10**(att/20)**2)-1)
b =  5*df
a =  4*df**2
d = b**2-4*a*c
print(a,b,c,d)

if d > 0: x = (-b+ (b**2-4*a*c)**0.5 )/(2*a)
else    : x = (-b+ (b**2+4*a*c)**0.5 )/(2*a)

Q1 = x**0.5
Q2 = Q1 * 2

print('$$Q1 =  {:1.2f}$$'.format(Q1))
print('$$Q2 =  {:1.2f}$$'.format(Q2))


# Simple sintonizado salida
Qo   = 200
rext = r22/2

L = (1/Q2-1/Qo)*(rext)/wo

Cs = 1/(L*wo**2) 
print('$$Cs= {:1.2e} F$$'.format(Cs))

Qm  = ((r22/rl) - 1)**0.5
C4  = 1/(rl*wo*Qm)

C4p = C4/(1+1/Qm**2)
C3  = Cs - C4p

print('$$L= {:1.2e} Hy$$'.format(L))

print('$$C3= {:1.2e} F$$'.format(C3))

print('$$C4= {:1.2e} F$$'.format(C4))

# Simple sintonizado entrada
rext1 = L*wo/(1/Q1-1/Qo)
r11t  = rext1*2
rgt   = rext1*2

Qm2   = r11t*wo*Cs
Nc    = (r11t/r11)**0.5
Qm1   = Qm2/Nc

print('$$Qm2 = {:1.2f}$$'.format(Qm2))
print('$$Qm1 = {:1.2f}$$'.format(Qm1))

C2  = Cs*Nc
C1  = Cs*Nc/(Nc-1)
print('$$C1= {:1.2e} F$$'.format(C1))

print('$$C2= {:1.2e} F$$'.format(C2))

N1 = N * (rg/rgt)**0.5

print('$$N1 = {:1.2f}$$'.format(N1))
```

La transferencia del filtro Butterworth:

$|A| = \frac{1}{\sqrt{1 +  (\frac{f_{ca} - f_o}{f_c - f_o}})^{2\cdot3} } =  \frac{1}{10^{\frac{60}{20}}}$

$f_{ca} - f_o = 50 KHz$

$(\frac{50 KHz}{f_c - f_o})  =  10$

$f_c =  f_o - \frac{50 KHz}{10}$

$f_{mB} = f_o + \frac{BW}{2} cos(\frac{2m-1}{2n}\pi)$

$f_1 =  f_o + \frac{BW}{2}  \cdot cos(30)$

$$Q2 =  45.50$$


Tercer polo:

$f_3 =  f_o + \frac{BW}{2}  \cdot cos(150)$
