## 1 1. Obtener la expresión de la transferencia de tensión (módulo y fase) ($\frac{V_o}{V_{osc}}$)


![ex 2.13](fig/BeleskasPMRF.png)

Suponiendo $R_L = Inf$.


$$v_1 = v_i - (g_{m} \cdot v_1) \cdot \frac{1}{SC} $$

$$v_1 = v_i \frac{1}{1+   \frac{g_{m}}{SC}} $$


$$v_o = v_1 - (g_{m} \cdot v_1) \cdot \frac{1}{SC} = v_1  (1 - \frac{g_{m}}{SC})$$

$$v_o = v_i \frac{1}{1+   \frac{g_{m}}{SC}}  (1 - \frac{g_{m}}{SC})$$


$$\frac{v_o}{v_i} =  \frac{SC - g_{m}}{SC + g_{m}}$$


$$|A(j\omega)| =  \sqrt{ \frac{(\omega C)^2  + g_{m}^2}{(\omega C)^2  + g_{m}^2}}= 1$$


$$\varphi (A(j\omega)) = arctg(-\frac{\omega C}{g_{m}}) - arctg(\frac{\omega C}{g_{m}})$$

$$\varphi (A(j\omega)) = -2 arctg(\frac{\omega C}{g_{m}})$$

$$\Delta \varphi = \frac{d \varphi }{d g_{m}} \Delta g_{m}$$

Llamando $g_{mq}$ a la trasconductancia del punto de polarización.  


$$\frac{d \varphi(g_{mq}) }{d g_{m}} = \frac{2 \omega C}{g_{mq}^2 + (\omega C)^2}$$

$$\Delta \varphi = \frac{2 \omega C}{g_{mq}^2 + (\omega C)^2} \Delta g_{m}$$


## 2. Obtener el valor de C que maximiza la sensibilidad del modulador $\frac{\Delta \Phi}{\Delta g_m}$

La variación de la fase depende de la variación de $g_m$, que puede ser calculada como:

$$\Delta \varphi = \frac{2 \omega C}{g_{mq}^2 + (\omega C)^2} \Delta g_{m}$$

Se busca maximizar esta variación.

$$f(x) = \frac{2 x}{a^2 + x^2}$$

Donde 

$x = \omega C$ 

$a = g_{mq}$

$$\frac{\partial f(x)}{\partial x} = \frac{2}{a^2 + x^2} - \frac{4 x^2}{(a^2 + x^2)^2}$$

$$\frac{\partial f(x)}{\partial x} = \frac{2 (a^2 + x^2)}{(a^2 + x^2)^2} - \frac{4 x^2}{(a^2 + x^2)^2}$$ 

$$\frac{\partial f(x)}{\partial x} = \frac{2 a^2 - 2 x^2}{(a^2 + x^2)^2}  = 0$$ 

Entonces 

$$x = a$$

$$\omega C = g_{mq}$$

### Calculo de $g_{mq}$

La transferencia del integrador de entrada para DC es unitaria. 

Entonces, $V_{GSq} = -2.5 V$

Siendo $g_m = \frac{-2 I_{DSS}}{V_C} (1-\frac{V_{GS}}{V_C})$

$g_{mq} = \frac{-2 I_{DSS}}{V_C} (1-\frac{V_{GSq}}{V_C})$

$$g_{mq} = 2.0 mS$$

Recordando

$\omega C = g_{mq}$

donde $\omega = 2 \pi f_c$

$C = \frac{g_{mq}}{2 \pi f_c}$

$$C = 3.183 \times 10^{-9} F$$

## 3. Determinar el valor de $R_1$ que produce un $\Delta f$ de $50Hz$ a la salida del modulador

Vamos a buscar como $V_{G1}$ produce variaciones en $\Delta f$. 

Para ello es necesario encontrar las ecuaciones que las vinculan. 

$$V_{G1} => \Delta V_{GS} => \Delta g_{m} =>  \Delta \varphi => \Delta f$$  

### $\Delta V_{GS} = |A_v| V_{G1}$

La transferencia del integrador a la frecuencia del mensaje ($1 KHz$)

$$A_v = -\frac{Z_2}{Z_1}$$ 

$$Z_1 = R_1$$

$$Z_2 = \frac{R_2}{1 + S C_2 R_2}$$

$$A_v = -\frac{R_2}{R_1} \frac{1}{1+S C_2 R_2}$$ 

$$A_v = -\frac{R_2}{R_1} \frac{1}{1+j \omega C_2 R_2}$$ 

$$|A_v| = \frac{R_2}{R_1} \frac{1}{\sqrt{(1+ (\omega C_2 R_2)^2}}$$ 

Entonces

$$\Delta v_{gs} = |A_v| V_{G1}$$

###  $\Delta g_{m}$ 

$$\Delta g_{m} = \frac{\partial g_m}{\partial V_{GS}} \Delta V_{GS}$$


$$\Delta g_{m} = -\frac{2 I_{dss}}{V_c^2} \Delta V_{GS}$$

###  $\Delta \varphi$

Del desarrollo en 1:

$$\Delta \varphi = \frac{1}{g_{mq}} \Delta g_{m}$$


###  Relación entre $\Delta \varphi$ y $\Delta f$


$$\omega_i(t) = \omega_c + \Delta \omega(t)$$

$$\omega_i(t) = \omega_c + \Delta \omega \cdot \cos(\omega_m t)$$


$$\varphi = \int \omega_i(t) dt$$ 

$$\varphi = \omega_c t +  \frac{\Delta \omega_m}{\omega_m} \cdot \sin(\omega_m t)$$ 

$$\Delta \varphi = \frac{\Delta \omega_m}{\omega_m} = \frac{\Delta f_m}{f_m}$$

### Calculando 

$\Delta \varphi = \frac{\Delta f_m}{f_m}$

$$\Delta \varphi = 0.005$$

$\Delta g_{m} = \frac{\Delta \varphi}{2} g_{mq}$

$$\Delta g_m = 50 \mu S$$


$\Delta g_{m} = -\frac{2 I_{dss}}{V_c^2} \Delta V_{GS}$

$$\Delta V_{GS} = -62.5 mV$$


$|A_v| = \frac{R_2}{R_1} \frac{1}{\sqrt{(1+ (\omega C_2 R_2)^2}}$

$|A_v| = \frac{\Delta V_{GS}}{V_{G1}}$

$$R1 = 2.51 k \Omega$$


## 4. Obtener la potencia de salida a la frecuencia de canal

La entrada de $v_{RF} = 10 mV cos(2 \pi f_{RF})$, que corresponde a la salida del PLL, se asume de pequeña señal por su excursión. 

La señal de oscilador local produce la variación del punto de funcionamiento del dispositivo, d

Para obtener la potencia de salida a la frecuencia de canal es necesario conocer como varia la transconductancia $g_m$ en función de $V_{OL}$

Siendo 

$g_m = \frac{-2 I_{DSS}}{V_C} (1 - \frac{V_{GS}}{V_C})$

$g_m = g_{mo} (1 - \frac{V_{GS}}{V_C})$

Donde 

$V_{GS}(t) = V_{pol} + V_{OL} Square(2 \pi f_{OL} t)$

$Square()$ corresponde a una señal cuadrada.

Podemos escribir los máximos y minimos como:

$V_{GSMAX} = V_{pol} + V_{OL}$

$V_{GSMAX} = -0.5V$

$V_{GSMIN} = V_{pol} + V_{OL}$

$V_{GSMIN} = -2.5V$


$g_m(t) = g_{mq} + g_{mp} Square(2 \pi f_{OL} t)$

Donde 

$g_{mq}$ es la transconductancia en el punto de operación 

$g_{mp}$ es la transconductancia que cambia debido a $V_{OL}$

Estos valores pueden ser calculados mediante los máximos y minimos que produce $V_{OL}$


$g_{mMAX} = g_{mo} (1 - \frac{V_{GSMAX}}{V_C})$

$g_{mMAX} = 3.6 mS$

$g_{mMIN} = g_{mo} (1 - \frac{V_{GSMIN}}{V_C})$

$g_{mMIN} = 2.0 mS$


$g_m(t) = 2.8 mS + 0.8 mS Square(2 \pi f_{OL} t)$


Que tambien podemos escribir como

$g_m(\theta) = 2.8 mS + 0.8 mS Square(\theta)$


### Serie de Fourier trigonometrica

Luego la serie de Fourier asociada a $g_m(t)$ es:

$g_{m}(\theta) = \frac{a_0}{2} + \sum_{n=1}^\infty [ a_n cos(\theta n) + b_n sin(\theta n)]$

Donde $a_0$, $a_n$ y $b_n$ son los coeficientes de Fourier que toman los valores:

$a_0 = \frac{2}{2 \pi} \int_{-\pi}^{\pi}  g_{m}(\theta) dt$

$a_n = \frac{2}{2 \pi} \int_{-\pi}^{\pi}  g_{m}(\theta) cos(\theta n) d\theta$

$b_n = \frac{2}{2 \pi} \int_{-\pi}^{\pi}  g_{m}(\theta) sin(\theta n) d\theta$

Donde encontramos que para una señal cuadrada (Square) con amplitud $A$ pico a pico:

$a_n = \frac{2 A}{\pi n} sin( \frac{\pi}{2} n)$

$a_1 = \frac{2 A}{\pi}$

$a_2 = 0$

$a_3 = \frac{-2 A}{\pi 3}$

$a_4 = 0$

$a_5 = \frac{2 A}{\pi 5}$

$g_m(\theta) = 2.8 mS + 1.01 mS cos(\theta) - 0.34 mS cos(3 \theta) + 0.2 mS cos(5 \theta)$

$g_m(t) = 2.8 mS + 1.01 mS cos(2 \pi f_{OL} t) - 0.34 mS cos(2 \pi 3 f_{OL} t) + 0.2 mS cos(2 \pi 5 f_{OL} t)$


La corriente de salida que tiene producto en la frecuencia deseada con $v_{RF}$ es la de $3 f_{OL} = 1 MHz$.

$i_D = g_{m3} v_{RF}$

$i_D = 0.34 mS cos(2 \pi 3 f_{OL}) 10 mV cos(2 \pi f_{RF})$

$i_D = \frac{0.34 mS 10 mV}{2} cos(2 \pi (f_{RF}+3f_{OL}))$

$i_D = 1.7 \mu A$

$v_o = i_D R_{tot}$

$v_o = 8.5 mV$

Donde podemos llamar $g_{mc}$ (transconductancia de conversión)

$g_{mc} = \frac{0.34 mS}{2}$

La potencia entonces sobre la carga 

$P_{RL} = \frac{i_{D}^2}{8} R_L$

$P_{RL} = 14.45 nW$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy                  # http://scipy.org/
from scipy import signal
%matplotlib inline

from scipy.fftpack import fft, fftfreq
from time import time


# Calculo de Re
# Datos

#FET
Idss = 10e-3
Vc   = -5
Vg   =  0


#integrador
#R1   = 4.7e3
R2  = 10e3
C2  = 100e-9
fmod = 1e3 
vmod = 0.1

#Salida
fc = 100e3
Df = 50


# gm
def gmVgs(Vgs, Idss, vc):
    return -2 * Idss/Vc *(1-(Vgs/Vc))


Vgso = -2.5   
print('V_GSq= {:1.3f} V'.format(Vgso))

gmq   = gmVgs(Vgso,Idss,Vc)
print('$g_mq = {:1.3f} mS$'.format(gmq/1e-3))


C = gmq/(2*np.pi*fc)
print('$C = {:1.3e} F$'.format(C))

```

```python
Dphi = Df/fmod
print('$\ Delta \ varphi = {:1.2e}$'.format(Dphi))


Dgm  = Dphi/2 * gmq
print('$\ Delta g_m = {:1.2e} S$'.format(Dgm))

DVgs  = Dgm *(Vc**2/(-2 * Idss))
print('$\ Delta V_GS = {:1.2e} V$'.format(DVgs))


R1 = -R2/( (DVgs/vmod) * (1+(2*np.pi*fmod*R2*C2)**2)**0.5)
print('$R1 = {:1.2e}$'.format(R1))

```

```python
def idi(v, Idss = 10e-3, vc=-5):
    i =  0.              if (v < vc)                else 0
    i = Idss*(1-v/vc)**2 if (v >= vc  and v < 0)    else i
    i = Idss             if (v >= 0 )               else i
    return i


def gmi(v, Idss = 10e-3, vc=-5):
    i =  0.              if (v < vc)                else 0
    i = -2*Idss/vc*(1-v/vc) if (v >= vc  and v < 0)    else i
    i = Idss             if (v >= 0 )               else i
    return i


Idss = 10e-3
vc   = -5

vgsa = np.arange(vc,0,0.01)
idsa = np.array([idi(vgs, Idss, vc) for vgs in vgsa])
gma  = np.array([gmi(vgs, Idss, vc) for vgs in vgsa])


plt.plot(vgsa,idsa,'b.')
#plt.axis([-5, 15, -1, 30])
plt.grid()
plt.ylabel('ID(A)')
plt.xlabel('VGS(V)')
plt.show()


plt.plot(vgsa,gma,'b.')
#plt.axis([-5, 15, -1, 30])
plt.grid()
plt.ylabel('ID(A)')
plt.xlabel('VGS(V)')
plt.show()
```
