Repasemos uno por uno, desde los más fáciles de entender hasta los temas más complejos.
Ganar compresión

En un amplificador ideal, la ganancia debe ser constante sin importar cuál sea la potencia de entrada. Por supuesto, en la vida real, existe un límite en la cantidad de energía que podemos inyectar en nuestro amplificador antes de que la salida se desvíe de lo esperado. A un cierto nivel de potencia de entrada, la ganancia comenzará a disminuir.

Una cifra común para evaluar este fenómeno es el punto de compresión de 1 db. Es la potencia de salida a la que la ganancia real se desvía de la ganancia esperada en 1dB.  Esta cifra proporciona la mayor cantidad de potencia que podemos esperar de nuestro amplificador sin tener demasiada falta de linealidad.

## Distorsión armónica

Un amplificador ideal es lineal: su salida es simplemente una función lineal de su entrada. Ninguna otra frecuencia que no sea la presente en la entrada debería encontrarse en la salida.  Desafortunadamente, todos los amplificadores son no lineales hasta cierto punto, y en su salida encontraremos múltiplos de las frecuencias de entrada además de las frecuencias de entrada. Por ejemplo, si la señal de entrada es una onda sinusoidal pura, entonces en un osciloscopio (dominio del tiempo), esto se manifestaría como una onda sinusoidal distorsionada. En un analizador de espectro (dominio de frecuencia), veríamos energía en frecuencias distintas a la fundamental.

Matemáticamente, podemos medir esta distorsión con la THD - Relación de distorsión armónica total, definida por:

$THD = \frac{\sum_{n = 1}^{\infty} F_n}{F_1}$

Donde $F_n$ el poder de la $N-armónico$, y $F_1$ siendo la potencia de la fundamental.

Cuanto menor sea la THD, mejor.

##  Ruido

Se pueden dedicar libros enteros al ruido en los circuitos electrónicos. Pero por ahora, intentemos mantenerlo simple. Podemos identificar dos tipos diferentes de ruido: ruido interno y ruido externo.

El ruido externo proviene del exterior de nuestro circuito electrónico e incluye elementos como radiación cósmica, interferencia y ruido atmosférico. No hay mucho que podamos hacer con respecto a estas fuentes de ruido y tenemos que lidiar con ellas.

El ruido interno, sin embargo, se genera dentro de nuestro circuito. ¡Esto es algo que podemos controlar de alguna manera! No profundizaremos demasiado en los orígenes físicos del ruido. Es importante saber que existe y que el ruido se convertirá en un factor determinante en el rendimiento del receptor. Por ahora, solo tenga en cuenta que una gran parte del ruido interno generado dentro del circuito se debe al inevitable movimiento aleatorio de los electrones dentro de su circuito. Esto se vuelve más importante cuanto más alta es la temperatura.

Es posible cuantificar el ruido con el  factor de ruido F, o factor de ruido NF (en DB):

$F = \frac {N_{out}}{N_{in} G}$

$NF = 10log(F)$


Donde $N_{out}$ es la potencia de ruido de salida, $N_{in}$ siendo la potencia de ruido de entrada, y G es la ganancia del amplificador.

La figura de ruido NF se puede encontrar en la hoja de datos de un dispositivo. Sin embargo, deben tomarse con un grano de sal: la cifra que encontrará en la hoja de datos a menudo representa el mejor de los casos.


## IMD: distorsión de intermodulación

Este término aparecerá muy a menudo no solo en amplificadores, sino en casi todos los componentes básicos de la radio. IMD caracteriza cómo se comporta un circuito, amplificador o no, cuando se aplican 2 señales de diferentes frecuencias en su puerto de entrada.

En un amplificador, las salidas deseadas son las frecuencias de la señal amplificada f_1y f_2. Sin embargo, para cualquier circuito, siempre que se aplican dos o más frecuencias diferentes a su puerto de entrada, ocurre un fenómeno conocido como  intermodulación  : no solo encontraremos las frecuencias de entrada en la salida, sino que también tendremos la suma y la diferencia de estas frecuencias. .  Estas frecuencias se denominan productos IMD y ,  dependiendo de cómo se 'mezclen' las frecuencias de entrada, pueden ser productos IMD de segundo, tercer o mayor nivel:

    2ª productos para IMD: f_1 + f_2,f_1 - f_2
    3er productos para IMD: 2f_1 - f_2, 2f_2 -f_1, 2f_1 + f_2,2f_2 + f_1
    y así…

Al describir el comportamiento de IMD, decimos que los productos de n-ésimo orden están X-dB por debajo de la potencia de salida de la señal deseada.

IMD aumenta con la potencia de entrada. De hecho, los productos de enésimo orden aumentan en n dB por cada 1 dB de potencia de entrada.

Al extender la línea de salida ideal y las líneas de orden x, se hacen visibles puntos interesantes: la intersección entre estas dos líneas. Este punto se llama  punto de intersección . Este punto tiene coordenadas ( IP_ {XIN}, IP_ {XOUT}). Más específicamente, el punto de intercepción de tercer orden es de particular interés para nosotros y es una figura de mérito ampliamente utilizada tanto para amplificadores como para mezcladores. ¿Porqué es eso? Bueno, los productos de segundo orden están tan lejos de nuestro rango de frecuencias deseadas que se filtran fácilmente y no afectan mucho al rendimiento. Los productos de pedidos extraños sobre el pedido 3 simplemente no tienen suficiente potencia para molestarnos realmente, incluso si pueden acercarse bastante a nuestro rango de frecuencias deseado.

Sin embargo, los productos de tercer orden están dentro del rango de frecuencias deseadas (al menos los productos de diferencia lo están. Los productos de la suma están fuera del rango), por lo que son difíciles de filtrar Y también tienen suficiente potencia para afectar notablemente el rendimiento. Si bien el punto de intersección se puede trazar en el gráfico, este punto es ficticio y se usa solo como una figura de mérito. En realidad, es muy probable que nunca pueda ingresar suficiente energía para llegar a ese punto sin sobrecargar el circuito.

Cuanto más alto sea nuestro IP_3, más resistente a IMD es nuestro amplificador.

Conclusión

Para poder diseñar un buen amplificador, primero debemos saber qué caracteriza a un buen amplificador. Las siguientes características son muy importantes al considerar los amplificadores:
