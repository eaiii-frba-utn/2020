```python
import numpy as np
import matplotlib.pyplot as plt
import scipy                  # http://scipy.org/
from scipy import signal

from IPython.display import Image  
%matplotlib inline
params = {'legend.fontsize': 24,
          'figure.figsize': (10, 10),
         'axes.labelsize': 24,
         'axes.titlesize': 24,
         'xtick.labelsize':24,
         'ytick.labelsize':24}
plt.rcParams.update(params) 
```

```python
def vgsi(VGG,Vs,th):
    return VGG+Vs*cos(th)

def idi(v):
    i =  0.             if (v < 2.5) else 0
    i = (17.5/2.5)*(v-2.5) if (v >= 2.5  and v <  5.0) else i
    i = (17.5)+(v-5.0) if (v >= 5.0  and v < 7.5) else i
    i = (20) if (v >= 7.5  ) else i
    return i
vgs = np.arange(-0.,12.5,0.01)


ids = []
for v in vgs:
    ids.append(idi(v))

plt.plot(vgs,ids,'b.')
plt.axis([0, 15, 0, 22.5])
plt.grid()
plt.ylabel('ID(A)')
plt.xlabel('VGS(V)')
plt.show()
```
